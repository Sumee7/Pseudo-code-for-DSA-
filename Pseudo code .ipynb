Pseudo-code for a Linked List Implemented Doubly Circular Priority Queue

=> Using a linked list as its structural element, we implement a Doubly Circular Priority Queue which provides bidirectional traversal of the queue while taking element priority into account.

Data Structure Definition:

Node: These nodes contain data, priorities, and references to the following and previous nodes.

Priority Queue: 
The queue is ordered so that the first element is processed, and the queue loops back to the first element after the last element has been processed.




Node Strucuture 

Node {
    Data: any
    Priority: int
    Next: Node (points to the next node in the queue)
    Prev: Node (points to the previous node in the queue)
}


1. Priority Queue Strucuture

DoublyCircularPriorityQueue {
    Head: Node (points to the node with the highest priority)
    Tail: Node (points to the node with the lowest priority)
    
    isEmpty(): boolean
    insert(data, priority): void
    removeHighestPriority(): Node
    traverseForward(): void
    traverseBackward(): void
}


Pseudo code
1. Initialize the queue 
 function DoublyCircularPriorityQueue():
    Head ← null
    Tail ← null

2. Check if the queue is empty
function isEmpty():
    return Head = null

3. Insert a new node based on priority
function insert(data, priority):
    newNode ← Node(data, priority)
    
    if isEmpty():
        Head ← newNode
        Tail ← newNode
        newNode.Next ← newNode   // Circular link
        newNode.Prev ← newNode   // Circular link
    else:
        current ← Head
        while current != Tail and current.Priority ≤ priority:
            current ← current.Next
        
        if current = Head and priority < Head.Priority: // Insert at the front
            newNode.Next ← Head
            newNode.Prev ← Tail
            Head.Prev ← newNode
            Tail.Next ← newNode
            Head ← newNode
        else if current = Tail and priority ≥ Tail.Priority: // Insert at the end
            newNode.Next ← Head
            newNode.Prev ← Tail
            Tail.Next ← newNode
            Head.Prev ← newNode
            Tail ← newNode
        else: // Insert in the middle
            prevNode ← current.Prev
            newNode.Next ← current
            newNode.Prev ← prevNode
            prevNode.Next ← newNode
            current.Prev ← newNode

5. Traverse the Queue Forward

function traverseForward():
    if isEmpty():
        return
    
    current ← Head
    do:
        print("Data: ", current.Data, " Priority: ", current.Priority)
        current ← current.Next
    while current != Head

function traverseBackward():
    if isEmpty():
        return
    
    current ← Tail
    do:
        print("Data: ", current.Data, " Priority: ", current.Priority)
        current ← current.Prev
    while current != Tail
